# Filtering surfaces: extracting signals from the surfaces
In the previous notebook we saw that we can extract different components of the surfaces from a single rappresentation.
Richardson with his algorithm was actually filtering high-frequency of the coastline, even thogh its intention was only to measure the coastline the filtering was intrinsic to its approach. This is a very important concept to understand in surface metrology. The first **filtering** or better **convolution** of the signal of the surface is due to the instrument itself. The rolling envelope method, gives an idea of what's happaning when we are measuring a surface with a circular probe (e.g. ball), but can be also used as a processing method for filtering or extracting high-frequency signal. This particular system is known as __E__ system. In this tutorial we will have a look to **M** systems. Another set of processing tools used for extracting singlas from the acquired surface.

```python
X = np.linspace(0,1,400)
Amplitude = 0.2
Frequency = 10
Phase = 0
#We compute the sinusoidal function
y = Amplitude*np.sin(X*2*np.pi*Frequency+Phase)

Amplitude2 = 0.1
Frequency2 = 5
Phase2 = 0
#We compute the sinusoidal function
y2 = Amplitude2*np.sin(X*2*np.pi*Frequency2+Phase2)
```

```python
plt.plot(y)
plt.plot(y2)
plt.show()
plt.plot(y+y2)
```

```python
x = np.arange(0,8.001,0.001)
z1 = 1*np.sin(2*np.pi*x/2)
z2 = 0.5*np.sin(2*np.pi*x/0.5)
z = z1 + z2 
plt.plot(z)
```

```python
dx = 0.001
l = 8 
n = l/dx
zf = np.fft.fft(z)
xf = np.linspace(0.0, 1.0/(2.0*dx), n/2)
plt.stem(xf,zf[:int(n//2)])
plt.xlim(0,20)
```

```python
print (xf[zf[:int(n//2)] > 2])
```

```python
# Number of samplepoints
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = np.fft.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)

fig, ax = plt.subplots()
ax.plot(xf, 2.0/N * np.abs(yf[:N//2]),marker='x')
plt.show()
```
