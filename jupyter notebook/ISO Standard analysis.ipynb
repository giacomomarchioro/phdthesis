```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
import os
```

```python
g = pd.read_csv(os.path.join(os.getcwd(),"Data","ISO_TC 213.csv"))
```

```python
g
```

```python
g['Status'].value_counts().plot(kind='barh')
plt.ylabel("Number of standards")
```

```python
stat = g['Status'].value_counts()
stat
```

```python
op = g.price_CF.sum()
print( "Overall cost of published standard %s CF %s Euro" %(op,op*0.91))
```

```python
g['Number of pages'].sum()
```

```python
g['year'] = (np.where(g['Publication date'].str.contains('-'),
                  g['Publication date'].str.split('-').str[0],
                  g['Publication date']))
```

```python
gs = g.sort_values(by=['year'])
years = gs['year'].value_counts(sort=False)
plt.rcParams['figure.figsize'] = (9,3)
```

```python
years[sorted(years.index)].plot(kind='bar',color='gray')
plt.ylabel('Number of standards')
plt.rcParams['svg.fonttype'] = 'none'
plt.tight_layout()
plt.savefig('ISO.pdf')
```
